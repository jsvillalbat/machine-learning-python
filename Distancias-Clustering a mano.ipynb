## Distancias

```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
data = pd.read_csv('python-ml-course-master/datasets/movies/movies.csv',sep = ';')
data
```

```python
movies = data.columns.values.tolist()[1:]
movies
```

Si se quiere normalizar, lo cual se hace cuando los valores entre columnas tienen rangos diferentes se realiza de la siguiente forma:
* dato normalizado = $ \frac{data\ original - data\ con\ menor\ valor\ de\ la\ columna}{dato\ con\ mayor\ valor\ de\ la\ columna - dato\ con\ menor\ valor\ de\ la\ columna}$  
En este caso de las peliculas todas las columas va de 0 a 10 entonces no es necesario normalizar

```python
dd1 = distance_matrix(data[movies],data[movies],p=1) #Distancia Manhatan
dd2 = distance_matrix(data[movies],data[movies],p=2) #distancia euclidea
dd10 = distance_matrix(data[movies],data[movies],p=10) # Al aumentar el p se disminuye la distancia parecido a una norma LP
```

```python
def dm_to_df(dd,col_name):
    import pandas as pd
    return pd.DataFrame(dd,index=col_name,columns=col_name)
```

```python
dm_to_df(dd1,data['user_id'])
```

```python
dm_to_df(dd2,data['user_id'])
```

```python
dm_to_df(dd10,data['user_id'])
```

### Clustering Jerarquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
Z = linkage(data[movies],'average') #Utilizando un enlace promedio
plt.figure(figsize=(25,10))
plt.title('Dendrograma jerarquico para el Clustering')
plt.xlabel('Id de los usuarios de Netflix')
plt.ylabel('Distancia')
dendrogram(Z,leaf_rotation=90.,leaf_font_size=17.0)
plt.show()
```

```python
Z = linkage(data[movies],'ward') #Utilizando un enlace  de ward
plt.figure(figsize=(25,10))
plt.title('Dendrograma jerarquico para el Clustering')
plt.xlabel('Id de los usuarios de Netflix')
plt.ylabel('Distancia')
dendrogram(Z,leaf_rotation=90.,leaf_font_size=17.0)
plt.show()
```

```python
Z = linkage(data[movies],'complete') #Utilizando un enlace completo
plt.figure(figsize=(25,10))
plt.title('Dendrograma jerarquico para el Clustering')
plt.xlabel('Id de los usuarios de Netflix')
plt.ylabel('Distancia')
dendrogram(Z,leaf_rotation=90.,leaf_font_size=17.0)
plt.show()
```

```python
Z = linkage(data[movies],'single') #Utilizando un enlace simple
plt.figure(figsize=(25,10))
plt.title('Dendrograma jerarquico para el Clustering')
plt.xlabel('Id de los usuarios de Netflix')
plt.ylabel('Distancia')
dendrogram(Z,leaf_rotation=90.,leaf_font_size=17.0)
plt.show()
```

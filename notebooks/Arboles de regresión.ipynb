### Árboles de regresión

```python
import pandas as pd
```

```python
data = pd.read_csv('python-ml-course-master/datasets/boston/Boston.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
colnames = data.columns.values.tolist()
predictors = colnames[:13]
target = colnames[13]
X = data[predictors]
Y = data[target]
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regtree = DecisionTreeRegressor(min_samples_split=30,min_samples_leaf=10,random_state=0)
```

```python
regtree.fit(X,Y)
```

```python
preds = regtree.predict(data[predictors])
```

```python
data['preds'] = preds
```

```python
data[['preds','medv']]
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
cv = KFold(n_splits= 10, shuffle=True, random_state=1)
```

```python
scores = cross_val_score(regtree,X,Y,scoring='neg_mean_squared_error',cv = cv, n_jobs=1)
```

```python
score = np.mean(scores)
score
```

```python
list(zip(predictors,regtree.feature_importances_)) #Nivel de importancia de cada una de las variables
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor(n_jobs=2, oob_score=True,n_estimators=10000) # 10000 árbolres
forest.fit(X,Y)
```

```python
data['rforest_pred']= forest.oob_prediction_
```

```python
data[['rforest_pred','medv']]
```

```python
data['rforest_error2'] = (data['rforest_pred']-data['medv'])**2
sum(data['rforest_error2'])/len(data)
```

```python
forest.oob_score_
```

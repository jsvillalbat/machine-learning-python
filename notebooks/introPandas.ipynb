# Carga de datos con la funcion open
##### Últil si archivos que contiene gran informacion del orden de GB se quieren cargar, si se hace con pandas puede llenar toda la memoria, y con open se puede abrir linea por linea y eliminando la anterior para liberar RAM

```python
data3 = open('python-ml-course-master/datasets/customer-churn-model/Customer Churn Model.txt','r') #Modo read w (write)
```

```python
cols = data3.readline().strip().split(',') #Strip elima espacios en blnaco
n_cols = len(cols)
```

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
for line in data3:
    values = line.strip().split(',')
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter+=1
    
print('El dataset tiene %d filas y %d columnas'%(counter,n_cols))
```

```python
df3 = pd.DataFrame(main_dict) #Carga manual desde un archivo de texto separado por comas
df3.head()
```

## Lectura y escritura de ficheros

```python
mainpath = 'python-ml-course-master/datasets'
infile =mainpath +'/'+'customer-churn-model/Customer Churn Model.txt'
outfile = mainpath + '/'+ 'customer-churn-model/Tab Customer Churn Model.txt'
```

```python
with open(infile,'r') as infile1:
    with open(outfile,'w') as outfile1:
        for line in infile1:
            fields = line.strip().split(',')
            outfile1.write('\t'.join(fields)) #Los campos cambiaron de estan separados por coma a TAB('/t')
            outfile1.write('\n')
```

```python
df4 = pd.read_csv(outfile,sep = '\t')
df4.head()
```

### Leer datos desde una URL

```python
medals_url ='http://winterolympicsmedals.com/medals.csv'
```

```python
medals_data = pd.read_csv(medals_url)
```

```python
medals_data.head()
```

```python
import csv
import urllib3
```

```python
http = urllib3.PoolManager()
r = http.request('GET',medals_url)
r.status
response= r.data
```

```python
cadena = response.decode('utf-8')
```

```python
lineas = cadena.split('\n')
colum_names = lineas[0].split(',')

second_dict = {}
for col in colum_names:
    second_dict[col] = []
contador = 0

for line in lineas[1:]:
    values = line.strip().split(',')
    for i in range(len(colum_names)):
        second_dict[colum_names[i]].append(values[i])
    contador+=1
    
print('El dataset tiene %d filas y %d columnas'%(contador,len(colum_names)))
```

```python
medal_df = pd.DataFrame(second_dict)
medal_df.head()
```

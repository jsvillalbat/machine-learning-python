### El m√©todo del codo y elf ector de la silueta del clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
x1 = np.array([3,1,1,2,1,6,6,6,5,6,7,8,9,8,9,9,8])
x2 = np.array([5,4,5,6,5,8,6,7,6,7,1,2,1,2,3,2,3])
X = np.array(list(zip(x1,x2))).reshape(len(x1),2)
```

```python
plt.xlim([0,10])
plt.ylim([0,10])
plt.title('Dataset a clasificar')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x1,x2)
plt.show()
```

```python
max_k = 10 #maximo numero de clusters
K = range(1,max_k)
ssw = []
color_palette = [plt.cm.Spectral(float(i)/max_k) for i in K]
centroid = [sum(X)/len(X) for i in K]
sst = sum(np.min(cdist(X,centroid,'euclidean'),axis = 1))

for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    centers = pd.DataFrame(kmeanModel.cluster_centers_)
    labels = kmeanModel.labels_
    ssw_k = sum(np.min(cdist(X,kmeanModel.cluster_centers_,'euclidean'),axis = 1))
    ssw.append(ssw_k)
    
    label_color = [color_palette[i] for i in labels]
    
    ## Fabricar una silueta para cada cluster
    #Por segurida no hacemos silueta si j = 1, o k = len(X)
    if 1<k<len(X):
        fig,(axis1,axis2) = plt.subplots(1,2)
        fig.set_size_inches(20,8)
        
        #El primer subplot contendr√° la silueta, que puede tener valores entre -1 y 1
        # En nuestro caso, ya controlamos que los valores est√°n entre -0.1 y 1
        axis1.set_xlim([-0.1,1.0])
        # EL numero de clsuters a insertar determinar√° el tama√±o de cada barra
        # El coeficiente (n_clusters+1)*10 ser√° el espacio en blanco que dejaremos entre siluetas individuales de cada cluster para separarlas
        axis1.set_ylim([0,len(X) + (k+1)*10])
        
        silhoutte_avg = silhouette_score(X,labels)
        print("* Para k = ",k," el promedio de la silueta es de: ",silhoutte_avg)
        sample_silhoutte_values = silhouette_samples(X,labels)
        
        y_lower = 10
        for i in range(k):
            #Agregamos la silueta del cluster k-√©simo
            ith_cluster_sv = sample_silhoutte_values[labels==i]
            print("   -Para i = ", i+1," la silueta del cluster vale: ", np.mean(ith_cluster_sv))
            #Ordenamos descendentemente las siluetas del i-√©simo
            ith_cluster_sv.sort()
            
            #Calculamos donde colocar la primera silueta en el eje vertical
            ith_cluster_size = ith_cluster_sv.shape[0] 
            y_upper = y_lower+ith_cluster_size
            
            #Elegimos el color del cluster
            color =  color_palette[i]
            #Pintamos la silueta del cluster i-√©simo
            axis1.fill_betweenx(np.arange(y_lower,y_upper),0,ith_cluster_sv,facecolor = color,alpha = 0.7)
            
            #Etiquetamos dicho cluster con el n√∫mero en el centro
            axis1.text(-0.05,y_lower+0.5*ith_cluster_size,str(i+1))
            
            #Calculamos el numero y_lower para el siguiente grafico del cluster
            y_lower = y_upper+10 # dejamos vacias 10 posiciones sin muestra
            
        axis1.set_title('Representaci√≥n de la silueta para k = %s' %str(k))
        axis1.set_xlabel('S(i)')
        axis1.set_ylabel('Id del cluster')
        
        # Fin de la representaci√≥n de la silueta
        
    # Plot de los k-means con los puntos respectivos
    plt.plot() #grafico vacio
    plt.xlim([0,10])
    plt.ylim([0,10])
    plt.title('Clustering para k = %s' %str(k))
    plt.scatter(x1,x2,c= label_color)
    plt.scatter(centers[0],centers[1],c = color_palette[k],marker = 'x')
    plt.show()
    
    
    
```

```python
## Representaci√≥n del codo
plt.plot(K,ssw,'bx-')
plt.xlabel('k')
plt.ylabel('SSw(k)')
plt.title('La t√©cnica del codo para encontrar el k √≥ptimo')
plt.show()
           
```

```python
## Representaci√≥n del codo normalizado
plt.plot(K,1-ssw/sst,'bx-')
plt.xlabel('k')
plt.ylabel('1-norm(SSw(k))')
plt.title('La t√©cnica del codo normalizado para encontrar el k √≥ptimo')
plt.show()
```
